In this notebook, I add new basis vectors from the single-cell RNAseq data contained in folders labeled by time point (E9, E13.5, E15.5, E17.5) inside the data folder. Each folder contains tSNE plots with labeled clusters, for different levels of resolution of the clustering algorithm. The data we are using comes from resolution 0.5, and the corresponding tSNE. I don't know what the clusters correspond to biologically, so I have used the same cluster labels as were used in those plots and in the accompanying tables.

# Imports and Data Loading

```python
import os
import pandas as pd
import numpy as np
import pickle
import lap_v2_py3 as lap_v2
import matplotlib.pyplot as plt
import matplotlib
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
from matplotlib.backends import backend_pdf as bpdf
matplotlib.style.use('ggplot')
%matplotlib inline

#Potentiall confusing projections:
#exceptions = ['E.13 Thyroid','E.9 Lung Prim Nkx+','E.9 Forebrain','E.8.25 Ectoderm','E.8.25 Foregut Endoderm']
#exceptions = ['E.9 Lung Prim Nkx+']
#exceptions = None

#Don't use clusters with too many zeros
exceptions = ['E9 cluster 3','E9 cluster 4', 'E9 cluster 5', 'E9 cluster 6', 'E9 cluster 7', 'E13 cluster 2', 'E13 cluster 3','E13 cluster 4', 'E15.5 cluster 3', 'E15.5 cluster 4','E17.5 cluster 2','E17.5 cluster 3','E17.5 cluster 4','E17.5 cluster 5']

#Folder with data:
folder = '../Data/'

f = open(folder+'master_table.dat','rb')
master_table, data_keys = pickle.load(f)
f.close()

master_table.dropna(inplace=True)

#Remove Nkx2-1
#master_table.drop(21869,axis=0,inplace=True)

def average_groups(data,name_dict):
    names_all = [item for sublist in name_dict.values() for item in sublist]
    reduced_data = data.drop(names_all,1)
    
    for name in name_dict:
        reduced_data = reduced_data.join(pd.DataFrame(data = data[name_dict[name]].mean(1), columns = [name]))

    return reduced_data

def pd_lap(names, exceptions = None, data = master_table, full = False):
    basis = data.drop(data_keys,1)
    
    if exceptions != None:
        reduced_basis = basis.drop(exceptions,1)
    else:
        reduced_basis = basis
        
    [a, A, eta] = lap_v2.lap(reduced_basis.values, data[names].values, full_output=True)

    projections = pd.DataFrame(data = a, index = reduced_basis.keys(), columns = names)
    if full:
        eta = pd.DataFrame(data = eta.T, index = data.index, columns = reduced_basis.keys())
        return projections,eta
    else:
        return projections
```

# Make projections

```python
data = master_table_filtered

projections = {}
expt_list = {
    'Transient Nkx2-1':['D1-1','D1-2','D7-1','D7-2','D7-3','D14-1','D14-2'],
    'BMP4+FGF2':['BMP4+FGF2 mCherry+ 1','BMP4+FGF2 mCherry+ 2','BMP4+FGF2 mCherry+ 3','BMP4+FGF2 mCherry- 1','BMP4+FGF2 mCherry- 2','BMP4+FGF2 mCherry- 3'],
    'BMP4+Wnt3A':['BMP4+Wnt3A mCherry+ 1','BMP4+Wnt3A mCherry+ 2','BMP4+Wnt3A mCherry+ 3','BMP4+Wnt3A mCherry- 1','BMP4+Wnt3A mCherry- 2','BMP4+Wnt3A mCherry- 3'],
    '3D EpCAM':['3D_EpCAM_2', '3D_EpCAM_3', '3D_EpCAM_4', '3D_mCherry-EpCAM_2','3D_mCherry-EpCAM_3', '3D_mCherry-EpCAM_4'],
    'Gelatin EpCAM':['Gelatin_mCherry-EpCAM_2','Gelatin_mCherry-EpCAM_3', 'Gelatin_mCherry-EpCAM_4']
}
first = True
for item in expt_list:
    if first:
        projections[item], eta = pd_lap(expt_list[item],data=data,full = True,exceptions=exceptions)
        first = False
    else:
        projections[item] = pd_lap(expt_list[item],data=data,exceptions=exceptions)
    
```

```python
for cell_type in ['E13 cluster 1','E.9 Lung Prim Nkx+']:
    fig,ax=plt.subplots(figsize=(12,24))
    eta = eta.reindex(master_table.index)
    names = ['3D_mCherry-EpCAM_2','3D_mCherry-EpCAM_3','3D_mCherry-EpCAM_4']
    genes = pd.DataFrame((master_table[names].values.T*eta[cell_type].values).T,index=master_table.index,columns=names).reindex(conv.index)
    genes.index = conv['symbol']
    genes = genes.sort_values(names[0],ascending=False)
    genes.iloc[:100].sort_values(names[0]).plot.barh(ax=ax,legend='reverse')

    ax.set_title('Top 100 Contributions to Projection of 3D EpCAM+/Nkx+ on '+cell_type)
    ax.set_xlabel('Contribution to Projection')
    pdf = bpdf.PdfPages('3DmCherry+'+cell_type+'.pdf')
    pdf.savefig(fig)
    pdf.close()
    plt.show()
```

# Imports and Data Loading

```python
import os
import pandas as pd
import numpy as np
import pickle
import lap_v2_py3 as lap_v2
import matplotlib.pyplot as plt
import matplotlib
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
from matplotlib.backends import backend_pdf as bpdf
matplotlib.style.use('ggplot')
%matplotlib inline

#Folder with data:
folder = '../Data/'

f = open(folder+'master_table.dat','rb')
master_table, data_keys = pickle.load(f)
f.close()

master_table.dropna(inplace=True)

#Remove Nkx2-1
#master_table.drop(21869,axis=0,inplace=True)

def average_groups(data,name_dict):
    names_all = [item for sublist in name_dict.values() for item in sublist]
    reduced_data = data.drop(names_all,1)
    
    for name in name_dict:
        reduced_data = reduced_data.join(pd.DataFrame(data = data[name_dict[name]].mean(1), columns = [name]))

    return reduced_data

def pd_lap(names, exceptions = None, data = master_table):
    basis = data.drop(data_keys,1)
    
    if exceptions != None:
        reduced_basis = basis.drop(exceptions,1)
    else:
        reduced_basis = basis
        
    [a, A, eta] = lap_v2.lap(reduced_basis.values, data[names].values, full_output=True)

    projections = pd.DataFrame(data = a, index = reduced_basis.keys(), columns = names)
    
    return projections

projections = {}
expt_list = {
    'Transient Nkx2-1':['D1-1','D1-2','D7-1','D7-2','D7-3','D14-1','D14-2'],
    'BMP4+FGF2':['BMP4+FGF2 mCherry+ 1','BMP4+FGF2 mCherry+ 2','BMP4+FGF2 mCherry+ 3','BMP4+FGF2 mCherry- 1','BMP4+FGF2 mCherry- 2','BMP4+FGF2 mCherry- 3'],
    'BMP4+Wnt3A':['BMP4+Wnt3A mCherry+ 1','BMP4+Wnt3A mCherry+ 2','BMP4+Wnt3A mCherry+ 3','BMP4+Wnt3A mCherry- 1','BMP4+Wnt3A mCherry- 2','BMP4+Wnt3A mCherry- 3'],
    '3D EpCAM':['3D_EpCAM_2', '3D_EpCAM_3', '3D_EpCAM_4', '3D_mCherry-EpCAM_2','3D_mCherry-EpCAM_3', '3D_mCherry-EpCAM_4'],
    'Gelatin EpCAM':['Gelatin_mCherry-EpCAM_2','Gelatin_mCherry-EpCAM_3', 'Gelatin_mCherry-EpCAM_4']
}
for item in expt_list:
    projections[item] = pd_lap(expt_list[item],data=master_table)
    
```

## Average groups and remove exceptions

```python
namedict = {'epithelial':['medullary thymic epithelial', 'terminal bronchiolar epithelial cells', 
                      'endothelial cell - gonad', 'endothelial cell - blood vessel',
                      'endothelial cell - lymphatic vessel'],
            'monocytes':['monocytes - classical', 'Granulocyte-Monocyte Progenitor (GMP)'],
            'adipose':['adipose - brown','adipose - white'],
            'killer cells':['natural killer cells','T Cell'],
            'intestine':['intestine - small','intestine - large'],
            'neuron':['neuron - glutamatergic','nsc'],
           'fibroblast':['fibroblast - thymus','fibroblast - skin','fibroblast - cardiac']}
table_fixed = average_groups(master_table,namedict)

exceptions=['Common Lymphoid Progenitor (CLP)','thyroid','lung','HSC','sperm','oocyte','mef']

projections_fixed = {}
for item in expt_list:
    projections_fixed[item] = pd_lap(expt_list[item],data=table_fixed,exceptions=exceptions)
```

# Visualize

## All data, bar plots

```python
fig,axs=plt.subplots(1,5,figsize=(32,32))
k = 0
for item in expt_list:
    projections[item].plot.barh(ax=axs[k],legend='reverse')
    axs[k].set_xlim((-0.2,0.7))
    axs[k].set_title(item)
    if k>0:
        axs[k].set_yticklabels(())
    k+=1

pdf = bpdf.PdfPages('bar-all.pdf')
pdf.savefig(fig)
pdf.close()
plt.show()
```

## 3D Plots, separate

```python
axes = ['E.13 Thyroid','E.9 Lung Prim Nkx+','E.8.25 Foregut Endoderm']

fig = plt.figure(figsize=(32,20))
#fig.subplots_adjust(hspace=0.4)
k = 0
axs = []
colors = sns.color_palette()

for item in expt_list:
    axs.append(fig.add_subplot(2,3,k+1,projection='3d'))
    #if item != 'Transient Nkx2-1':
    m = 0
    for point in projections[item].keys():
        axs[-1].scatter(projections[item][point].loc[axes[0]],projections[item][point].loc[axes[1]],projections[item][point].loc[axes[2]],label=point,color=colors[m],s=100)
        m+=1
    axs[-1].legend()
    #else:
    #    axs[-1].plot(projections[item].loc[axes[0]],projections[item].loc[axes[1]],projections[item].loc[axes[2]])
    axs[-1].set_xlabel(axes[0])
    axs[-1].set_ylabel(axes[1])
    axs[-1].set_zlabel(axes[2])
    axs[-1].set_xlim((0,0.5))
    axs[-1].set_ylim((0,0.5))
    axs[-1].set_zlim((0,0.5))
    axs[-1].set_title(item)
    
    k+=1

pdf = bpdf.PdfPages('3D-all.pdf')
pdf.savefig(fig)
pdf.close()
plt.show()
```

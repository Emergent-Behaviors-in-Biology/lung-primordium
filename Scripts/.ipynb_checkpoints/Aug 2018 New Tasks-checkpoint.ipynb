# Imports and Data Loading

```python
import os
import pandas as pd
import numpy as np
import pickle
import lap_v2_py3 as lap_v2
import matplotlib.pyplot as plt
import matplotlib
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
from matplotlib.backends import backend_pdf as bpdf
matplotlib.style.use('ggplot')
%matplotlib inline

#Potentiall confusing projections:
#exceptions = ['terminal bronchiolar epithelial cells','E.8.25 Foregut Endoderm']
exceptions = None

#Folder with data:
folder = '../Data/'

#Conversions
conv = pd.read_table(folder+'AnnotationTable_mogene10sttranscriptcluster.txt',index_col=2)
conv = conv[~conv.index.duplicated(keep=False)]

f = open(folder+'master_table.dat','rb')
master_table, data_keys = pickle.load(f)
f.close()

master_table.dropna(inplace=True)

#Remove Nkx2-1
#master_table.drop(21869,axis=0,inplace=True)

def average_groups(data,name_dict):
    names_all = [item for sublist in name_dict.values() for item in sublist]
    reduced_data = data.drop(names_all,1)
    
    for name in name_dict:
        reduced_data = reduced_data.join(pd.DataFrame(data = data[name_dict[name]].mean(1), columns = [name]))

    return reduced_data

def pd_lap(names, exceptions = None, data = master_table, full = False):
    basis = data.drop(data_keys,1)
    
    if exceptions != None:
        reduced_basis = basis.drop(exceptions,1)
    else:
        reduced_basis = basis
        
    [a, A, eta] = lap_v2.lap(reduced_basis.values, data[names].values, full_output=True)

    projections = pd.DataFrame(data = a, index = reduced_basis.keys(), columns = names)
    if full:
        eta = pd.DataFrame(data = eta.T, index = data.index, columns = reduced_basis.keys())
        return projections,eta
    else:
        return projections

projections = {}
expt_list = {
    'Transient Nkx2-1':['D1-1','D1-2','D7-1','D7-2','D7-3','D14-1','D14-2'],
    'BMP4+FGF2':['BMP4+FGF2 mCherry+ 1','BMP4+FGF2 mCherry+ 2','BMP4+FGF2 mCherry+ 3','BMP4+FGF2 mCherry- 1','BMP4+FGF2 mCherry- 2','BMP4+FGF2 mCherry- 3'],
    'BMP4+Wnt3A':['BMP4+Wnt3A mCherry+ 1','BMP4+Wnt3A mCherry+ 2','BMP4+Wnt3A mCherry+ 3','BMP4+Wnt3A mCherry- 1','BMP4+Wnt3A mCherry- 2','BMP4+Wnt3A mCherry- 3'],
    '3D EpCAM':['3D_EpCAM_2', '3D_EpCAM_3', '3D_EpCAM_4', '3D_mCherry-EpCAM_2','3D_mCherry-EpCAM_3', '3D_mCherry-EpCAM_4'],
    'Gelatin EpCAM':['Gelatin_mCherry-EpCAM_2','Gelatin_mCherry-EpCAM_3', 'Gelatin_mCherry-EpCAM_4']
}
first = True
for item in expt_list:
    if first:
        projections[item], eta = pd_lap(expt_list[item],data=master_table,full = True,exceptions=exceptions)
        first = False
    else:
        projections[item] = pd_lap(expt_list[item],data=master_table,exceptions=exceptions)
    
```

## Thyroid projection of 3D-mCherry-EpCam

```python
fig,ax=plt.subplots(figsize=(8,12))

genes = (master_table['3D_mCherry-EpCAM_2']*eta['E.13 Thyroid']).reindex(conv.index)
genes.index = conv['symbol']
genes = genes.sort_values(ascending=False)
genes.iloc[:50].sort_values().plot.barh(ax=ax)

ax.set_title('Top 20 Contributions to Projection of Double-Positive on Thyroid')
ax.set_xlabel('Contribution to Projection')
pdf = bpdf.PdfPages('mCherry-thyroid.pdf')
pdf.savefig(fig)
pdf.close()
plt.show()
```

## Difference of lung projection between mCherry+ and mCherry- 

```python
fig,ax=plt.subplots(figsize=(8,5))

genes = ((master_table['3D_EpCAM_2']-master_table['3D_mCherry-EpCAM_2'])*eta['E.9 Lung Prim Nkx+']).reindex(conv.index)
genes.index = conv['symbol']
genes = genes.sort_values(ascending=False)
genes.iloc[:20].sort_values().plot.barh(ax=ax)

ax.set_title('Top 20 Contributions to Superior Lung Projection of mCherry-')
ax.set_xlabel('Contribution to Projection Difference')

pdf = bpdf.PdfPages('mCherry-difference-lung.pdf')
pdf.savefig(fig)
pdf.close()
plt.show()
```

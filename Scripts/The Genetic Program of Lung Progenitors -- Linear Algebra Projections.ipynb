```python
import os
import pandas as pd
import numpy as np
import pickle
import lap_v2_py3 as lap_v2
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
from matplotlib.backends import backend_pdf as bpdf
matplotlib.style.use('ggplot')
%matplotlib inline

#Folder with data:
folder = '../Processed_Data/'

#Load processed data
#(Generated using )
f = open(folder+'master_table.dat','rb')
master_table, data_keys = pickle.load(f)
f.close()
master_table.dropna(inplace=True)

#Define projection function
def pd_lap(names, exceptions = None, data = None, full = False):
    basis = data.drop(data_keys,1)
    
    if exceptions != None:
        reduced_basis = basis.drop(exceptions,1)
    else:
        reduced_basis = basis
        
    [a, A, eta] = lap_v2.lap(reduced_basis.values, data[names].values, full_output=True)

    projections = pd.DataFrame(data = a, index = reduced_basis.keys(), columns = names)
    if full:
        eta = pd.DataFrame(data = eta.T, index = data.index, columns = reduced_basis.keys())
        return projections,eta
    else:
        return projections
```

# Make projections

## Make list of scenarios to run

```python
#Check what happens when different sets of basis vectors are excluded before projection
#In manuscript, we use all the basis types except for E.13.5 Hepatoblast
exceptions = {}
exceptions['single-cell'] = ['E9 cluster 0', 'E9 cluster 1', 'E9 cluster 2', 'E9 cluster 3',
       'E9 cluster 4', 'E9 cluster 5', 'E9 cluster 6', 'E9 cluster 7',
       'E13 cluster 0', 'E13 cluster 1', 'E13 cluster 2', 'E13 cluster 3',
       'E13 cluster 4', 'E15.5 cluster 0', 'E15.5 cluster 1',
       'E15.5 cluster 2', 'E15.5 cluster 3', 'E15.5 cluster 4',
       'E17.5 cluster 0', 'E17.5 cluster 1', 'E17.5 cluster 2',
       'E17.5 cluster 3', 'E17.5 cluster 4', 'E17.5 cluster 5']
exceptions['new files'] = ['E.13.5 Hepatoblast','E.16.5 Lung'] + exceptions['single-cell']
exceptions['hepatoblast'] = ['E.13.5 Hepatoblast']

#Organize test data by experiment. 
expt_list = {
    #Previous work (Dame, K. et al., Thyroid Progenitors Are Robustly Derived from Embryonic Stem Cells through Transient, Developmental Stage-Specific Overexpression of Nkx2-1. Stem Cell Rep 8, 216-225 (2017).)
    'Transient Nkx2-1':['D1-1','D1-2','D7-1','D7-2','D7-3','D14-1','D14-2','D14-3'],
    #First set of new experiments
    'BMP4+FGF2':['BMP4+FGF2 mCherry+ 1','BMP4+FGF2 mCherry+ 2','BMP4+FGF2 mCherry+ 3','BMP4+FGF2 mCherry- 1','BMP4+FGF2 mCherry- 2','BMP4+FGF2 mCherry- 3'],
    '2D Nkx2-1-':['BMP4+Wnt3A mCherry- 1','BMP4+Wnt3A mCherry- 2','BMP4+Wnt3A mCherry- 3'],
    '2D Nkx2-1+':['BMP4+Wnt3A mCherry+ 1','BMP4+Wnt3A mCherry+ 2','BMP4+Wnt3A mCherry+ 3'],
    #Second set of new experiments:
    '3D Nkx2-1+EpCAM+':['3D_EpCAM_2', '3D_EpCAM_3', '3D_EpCAM_4'],
    '3D Nkx2-1-EpCAM+':['3D_mCherry-EpCAM_2','3D_mCherry-EpCAM_3', '3D_mCherry-EpCAM_4'],
    '2D Nkx2-1+EpCAM+':['Gelatin_mCherry-EpCAM_2','Gelatin_mCherry-EpCAM_3', 'Gelatin_mCherry-EpCAM_4']
}
```

## Get projections

```python
#Choose which conditions to save to Excel
z_thresh_save = 1
ex_name_save = 'hepatoblast'

#Initialize dictionaries to hold all conditions
projections = {}
projections_full = {}

#Option to only use genes that are highly expressed in at least one sample
#In manuscript, we use z_thresh = 1
for z_thresh in [-5,1]:
    z_name = 'z > ' +str(z_thresh)
    projections[z_name] = {}
    
    #The manuscript data is ex_name=hepatoblast
    for ex_name in exceptions.keys():
        title = ex_name + ' excluded, ' + z_name
        
        data = master_table[(master_table.T > z_thresh).any()]
        projections[z_name][ex_name] = {}
        
        for item in expt_list:
            if ex_name == ex_name_save:
                if item == 'Gelatin EpCAM+':
                    if z_thresh == z_thresh_save:
                        #Save projection matrix eta just for the actual basis used in the manuscript
                        projections_full[item], eta = pd_lap(expt_list[item],data=data,full = True,exceptions=exceptions[ex_name])

            projections[z_name][ex_name][item] = pd_lap(expt_list[item],data=data,exceptions=exceptions[ex_name])

data = master_table[(master_table.T > z_thresh_save).any()]
pd_lap([item for key in expt_list.keys() for item in expt_list[key]],data=data,exceptions=exceptions[ex_name_save]).to_excel(folder+'projections.xlsx')
```

# Make bar plots

```python
ref_lineage = 'E.9 Lung Prim Nkx+'
ex_name = 'hepatoblast'
z_name = 'z > 1'

first = True
for expt_name in ['2D Nkx2-1+','2D Nkx2-1+EpCAM+','3D Nkx2-1+EpCAM+','3D Nkx2-1-EpCAM+']:
    results_new = pd.DataFrame(projections[z_name][ex_name][expt_name].loc[ref_lineage])
    results_new['Protocol'] = expt_name
    results_new = results_new.set_index('Protocol',append=True).swaplevel()
    if first:
        results = results_new
        first=False
    else:
        results = results.append(results_new)

results.groupby(level=0).mean().plot.bar()
plt.show()
```

# Look at all projections

```python
for z_name in projections.keys():
    for ex_name in exceptions.keys():
        title = ex_name + ' excluded, ' + z_name
        
        print('------------------------------------------')
        print(title)
        fig,axs=plt.subplots(2,4,figsize=(32,64))
        axs=axs.reshape(-1)
        k = 0
        for item in expt_list:
            projections[z_name][ex_name][item].plot.barh(ax=axs[k],legend='reverse')
            axs[k].set_xlim((-0.2,1))
            axs[k].set_title(item)
            if k>0:
                axs[k].set_yticklabels(())
            k+=1
        plt.show()
```

# Projections of randomized sequence data

## Shuffle the data

```python
#Shuffle the "query" data (keeping the "reference" data unchanged)
master_table_rand = master_table.copy()
for item in expt_list:
    for cond in expt_list[item]:
        master_table_rand[cond] = master_table[cond].sample(n=len(master_table),replace=False).values

#Run the original pipeline on the shuffled data
projections_rand = {}
for z_thresh in [-5,1]:
    z_name = 'z > ' +str(z_thresh)
    projections_rand[z_name] = {}
    
    for ex_name in exceptions.keys():
        title = ex_name + ' excluded, ' + z_name
        
        data = master_table_rand[(master_table_rand.T > z_thresh).any()]
        projections_rand[z_name][ex_name] = {}
        
        for item in expt_list:
            projections_rand[z_name][ex_name][item] = pd_lap(expt_list[item],data=data,exceptions=exceptions[ex_name])
```

## Visualize results

```python
# Look at the results
for z_name in projections_rand.keys():
    for ex_name in exceptions.keys():
        title = ex_name + ' excluded, ' + z_name
        
        print('------------------------------------------')
        print(title)
        fig,axs=plt.subplots(2,4,figsize=(32,64))
        axs=axs.reshape(-1)
        k = 0
        for item in expt_list:
            projections_rand[z_name][ex_name][item].plot.barh(ax=axs[k],legend='reverse')
            axs[k].set_xlim((-0.2,1))
            axs[k].set_title(item)
            if k>0:
                axs[k].set_yticklabels(())
            k+=1
        plt.show()
```
